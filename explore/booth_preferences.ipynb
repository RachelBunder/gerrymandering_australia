# Booth Preferences
This notebook explores the both preference data supplied by the Australian Electoral Commission (AEC)

```python
import pandas as pd
```

## Read in booth data

```python
booth_info_loc = 'data/20190518/GeneralPollingPlacesDownload-24310.csv'

try:
    booths = pd.read_csv(booth_info_loc,
                         skiprows=1)
except IOError:
    booths = pd.read_csv('https://results.aec.gov.au/24310/Website/Downloads/GeneralPollingPlacesDownload-24310.csv')
    booths.to_csv(booth_info_loc)
    booths = pd.read_csv(booth_info_loc,
                         skiprows=1)

booths.head()
```

```python
act_first_pref_loc = 'data/20190518/HouseStateFirstPrefsByPollingPlaceDownload-24310-ACT.csv'

try:
    act_first_pref = pd.read_csv(act_first_pref_loc,
                         skiprows=1)
except IOError:
    act_first_pref = pd.read_csv('https://results.aec.gov.au/24310/Website/Downloads/GeneralPollingPlacesDownload-24310.csv')
    act_first_pref.to_csv(act_first_pref_loc)
    act_first_pref = pd.read_csv(act_first_pref_loc,
                         skiprows=1)

act_first_pref.head()
```
